**COMBINING LINEAMENT DATASET **

---



```python
import geopandas as gpd
shapefile_path = '/content/lineament_Types.shp'
lineament = gpd.read_file(shapefile_path)

lineament
```

```python
import geopandas as gpd
shapefile_path = '/content/lineament_form.shp'
lineament1 = gpd.read_file(shapefile_path)

lineament1
```

```python

lineament_subset = lineament[['OBJECTID', 'LINEAMENT_']]
lineament1_subset = lineament1[['OBJECTID', 'LINEAMENT_']]

print("lineament subset:")
print(lineament_subset)

print("\nlineament1 subset:")
lineament1_subset
```

```python
import geopandas as gpd
import pandas as pd
from google.colab import files

columns_to_keep = ['OBJECTID', 'LINEAMENT_', 'geometry']

# Filter columns, ensuring 'geometry' is present
lineament_subset = lineament[lineament.columns.intersection(columns_to_keep)]
lineament1_subset = lineament1[lineament1.columns.intersection(columns_to_keep)]

print("lineament subset:")
print(lineament_subset)

print("\nlineament1 subset:")
print(lineament1_subset)

lineament_combined_df = pd.concat([lineament_subset, lineament1_subset], ignore_index=True)


if 'geometry' in lineament_combined_df.columns:
    lineament_combined = gpd.GeoDataFrame(lineament_combined_df, geometry='geometry')
else:

    print("Warning: 'geometry' column not found. Cannot create a GeoDataFrame for GeoJSON output.")

    lineament_combined = lineament_combined_df

print("\nCombined lineament and lineament1 subset:")
print(lineament_combined)

print("\nNull values in combined data:")
print(lineament_combined.isnull().sum())

output_geojson_path = 'lineament_combined.geojson'


if isinstance(lineament_combined, gpd.GeoDataFrame):
    lineament_combined.to_file(output_geojson_path, driver='GeoJSON')

    print(f"\nCombined lineament dataset saved to '{output_geojson_path}' in GeoJSON format.")

    !ls -l {output_geojson_path}

    # Download the file
    files.download(output_geojson_path)
else:
    print("\nCannot save to GeoJSON as the combined data is not a GeoDataFrame.")
```

**COMBINING MINERAL POINTS WITH POLYGON DATA**

```python
import geopandas as gpd
import pandas as pd
mineral_points = gpd.read_file("/content/mineralization.geojson")
mineral_polygons = gpd.read_file("/content/mineral_polygon.geojson")

mineral_polygons = mineral_polygons.to_crs(mineral_points.crs)

polygon_as_points = mineral_polygons.copy()
polygon_as_points['geometry'] = polygon_as_points.centroid

combined_points = pd.concat([mineral_points, polygon_as_points], ignore_index=True)

combined_points.to_file("combined_mineral_points.geojson", driver='GeoJSON')
```
